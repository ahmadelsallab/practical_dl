## 3- Our first model: quick start

We're going to use a <b>pre-trained</b> model, that is, a model created by some one else to solve a different problem. Instead of building a model from scratch to solve a similar problem, we'll use a model trained on ImageNet (1.2 million images and 1000 classes) as a starting point. The model is a Convolutional Neural Network (CNN), a type of Neural Network that builds state-of-the-art models for computer vision. We'll be learning all about CNNs during this course.

Fast.ai used <kbd>ResNet34</kbd> in their courses, but we will use instead <kbd>MobileNet</kbd> to run faster. For information, <kbd>resnet</kbd> won the 2015 ImageNet competition. ### 3.1- The data
There are 2 ways to load the dataset, either it fits into your memory and in that case you can open all the images and store them into an numpy array. But, in most of the cases, the dataset will be too big and you will have to read only a part of it at the time. This is called a **Batch**. A batch is then a sub-part of your dataset. The size of the batch depends on the size of your memory. It might be 64, 32, 16 or 8 images...

Keras manages all that for us:
- The <kbd>ImageDataGenerator</kbd> class define the transformation you want to apply on each of the sample it will open. It has, at least, to apply the transformation defined by the model we are going to use. Indeed, we are going to use a pretrained model, so we have to use exactly the same transformation than the one used during its training. Most of the time, it substract the mean, normalize and few other things. We will see later that it does much more, the so called <kbd>Data Augmentation</kbd>. We keep that for a next lesson.
- The function <kbd>flow_from_directory</kbd> load batches of data

```python
from keras.preprocessing.image import ImageDataGenerator
from keras.applications.mobilenet import MobileNet, preprocess_input

batch_size=16 # bigger is your memory, bigger the batch size will be

# The path of our training and validation dataset
train_data_dir = PATH/'subdataset/train'
validation_data_dir = PATH/'subdataset/valid'

# Our Image generator does only the transformation defined by the model we are going to use
# preprocess_input belong to "keras.applications.mobilenet", so this is exactly what MobileNet needs!
train_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)
test_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)

# We define our generators below. They will generate batches of images.
# In most of case, they are different but in this case, we use same parameters for both
train_generator = train_datagen.flow_from_directory(train_data_dir,
                                                    target_size=(sz, sz),
                                                    batch_size=batch_size)

validation_generator = test_datagen.flow_from_directory(validation_data_dir,
                                                        target_size=(sz, sz),
                                                        batch_size=256,
                                                       shuffle=False)
```

### 3.2- The model
The model we are going to use has been trained on ImageNet challenge. It performs images classification among 1000 classes! We do not need all that as we are going only to classify a Cat versus a Dog! However, it still makes sense because it is quite likely that these 1000 classes include Cats, Dogs and much more. So, what we need to do is somehow to "shrink its outputs" or "specialized its output" to Cat and Dog.
Again, Keras does that for us. When we load the model, we just need to define the property <kbd>include_top</kbd> to <kbd>False</kbd>. Put it at <kbd>True</kbd> would mean that we want to do exactly what the model has been trained for. 

Then, we will add our onw layer to perform our binary classification. There is no need now to understand the meaning of these few added layers, we will see that later.

```python
from keras.models import Model 
from keras.layers import Dense, GlobalAveragePooling2D

def Load_Model():
    # our model is based on MobileNet, but we load only the layers before the classification
    base_model = MobileNet(weights='imagenet', include_top=False, input_shape=(sz,sz,3))
    # This loop below is needed to tell to Keras that we don t want to retrain our base model
    for layer in base_model.layers:
            layer.trainable=False

    # below, we define our own layers to perform our binary classification Cats vs. Dogs
    NbClasses = 2
    x = base_model.output
    x = GlobalAveragePooling2D()(x)
    x = Dense(1024, activation='relu')(x)
    output_layer = Dense(NbClasses, activation='softmax', name='softmax')(x)
    model = Model(inputs=base_model.input, outputs=output_layer)

    #x = net.output
    #x = Flatten()(x)
    #x = Dropout(0.5)(x)
    return model

model = Load_Model()
```

There is this usefull function <kbd>model.summary()</kbd> that shows all the layer of the model, its inputs, outputs, parameters. In the last few lines, there is the total number of parameters of our model, the trainable ones and non-trainable ones. You can check that only 1 million of parameters among 4 millions will be trained. They correspond mainly to the final Dense layer.

```python
model.summary()
```

### 3.3- The Learning Process

Now, we need to define how things are going to be trained. First, we need a criteria that measure how good is our prediction compared to the ground truth. This is the <kbd>Loss</kbd>. Indeed, in most learning models, error is calculated as the difference between the actual output and the predicted output:

<img src="https://raw.githubusercontent.com/valeoai/tutorial-images/master/images/01_Classification_Introduction/Loss.gif" width="15%">

The function that is used to compute this error is known as Loss Function **J(w)**. Different loss functions will give different errors for the same prediction, and thus have a considerable effect on the performance of the model. There are different Loss function for the task you are dealing with, such as Cross Entropy for classification task or L1 loss (Sum of absolute differences) for regression tasks.

Then naturally, the main objective during the training phase is to reduce (minimize) the loss function's value with respect to the model's parameters by changing the weight vector values through different optimization methods. We need a process that will find the best parameters of our model (the previous 2 millions parameters we saw above) which gives a prediction as close as possible from the label. This is the role of the <kbd>optimizer</kbd> to perform this task. 

Optimization functions usually calculate the gradient i.e. the partial derivative of loss function with respect to weights, and the weights are modified in the opposite direction of the calculated gradient. This cycle is repeated until we reach the minima of loss function.

One of the most popular optimization algorithm is called [<b>The Gradient Descent</k>](http://wiki.fast.ai/index.php/Gradient_Descent). It is used to minimize the Loss function by iteratively moving in the direction of steepest descent as defined by the negative of the gradient. It is a step by step approach where for each step, the gradient is computed. The size of these steps is called the learning rate. With a high learning rate we can cover more ground each step, but we risk overshooting the lowest point since the slope of the hill is constantly changing. With a very low learning rate, we can confidently move in the direction of the negative gradient since we are recalculating it so frequently. A low learning rate is more precise, but calculating the gradient is time-consuming, so it will take us a very long time to get to the bottom.

<img src="https://raw.githubusercontent.com/valeoai/tutorial-images/master/images/01_Classification_Introduction/GradientDescente.png" width="50%">



Once all settings are configured, we need to compile our model. This is specific to Keras. You just need to call <kbd>model.compile</kbd> with all previously defined parameters.

```python
model.compile(optimizer='SGD',loss='binary_crossentropy', metrics=['accuracy'])
```

This is now time to train our model, using the optimizer defined, on our dataset (Batches). For that, we just need to call <kbd>model.fit_generator</kbd>

Our optimizer run on an iterative process. So, we have also to define how many times it will see the entire dataset to improve iteratively the model parameters. This is called the number of <kbd>epoch</kbd>. 10 epoches means that the optimizer will go through 10 times the whole dataset

```python
epochs = 3
model.fit_generator(train_generator, epochs=epochs,
                    validation_data=validation_generator,
                    verbose=1)
```

Some few details about the data outputted by <kbd>fit_generator</kbd>. First, as we defined, the process went through 3 iterations (our number of epochs).
So, 3 lines are displayed, each contains:
- the total number of bacthes: 229 = number of samples(3656) / batch size (16)
- the time it took to go through is entire training set: 50 sec (on a 1080ti)
- the 2 following numbers are the Loss and Accuracy of the training dataset
- the 2 last numbers are teh Loss and Accuracy on validatation dataset.

You might have seen that during one epoch, the loss and accuracy are recalculed for each bach.

### Preprocessing function

We need to stop for a while and really understand why image has to be preprocess?

Images are composed of pixel for which values are between [0-255]. Most of signal processing and Machine Learning algorithms prefer data between [0-1] to avoid numerical overflows. There are also specific reasons for Deep Learning where some activation functions are not compatible with all kinds of input.

Furthermore, an RGB image is composed of 3 channels. We need to make sure that pixels from all these channels are in the same range of values. Otherwise, it will have some performances impact during our training such as lowest values would be discarded.

To make it short, we can not feed a Deep Neural Network with raw data, data has to be prepared and we call that step <b>Data Preprocessing</b>.

There are 2 main types of preprocessing:
- <b>Mean subtraction</b> is the most common form of preprocessing. It involves subtracting the image mean to all the pixels, and has the geometric interpretation of centering the cloud of data around the origin along every dimension. 

- <b>Normalization</b> refers to normalizing the data dimensions so that they are of approximately the same scale. There are two common ways of achieving this normalization. One is to divide each dimension by its standard deviation, once it has been zero-centered: (X /= np.std(X, axis = 0)). Another form of this preprocessing normalizes each dimension so that the min and max along the dimension is -1 and 1 respectively. 

<img src="https://raw.githubusercontent.com/valeoai/tutorial-images/master/images/01_Classification_Introduction/preprocess.jpeg" width="60%">

All the statistics for the prepocessing are computed on the entire training set.

More information on [CS231N course](http://cs231n.github.io/neural-networks-2/#datapre)

Back to our classification problem, we have used above a pretrained model from someone else, MobileNet here.
Those people have trained their model on their own dataset. As a consequence, the preprocessing has been defined by them and the pretrained model will work only if tested data are prepared in a similar way the model has been trained for.

That's why in Keras, when you load a model <kbd>from keras.applications.mobilenet import MobileNet, preprocess_input</kbd>, you load the preproocessing too!

The preprocessing function is provided to the <kbd>ImageDataGenerator</kbd> so that it knows how to prepare data for the model.

### Looking at best predictions and worse errors 


```python
def rand_by_mask(mask): 
    return np.random.choice(np.where(mask)[0], 2, replace=False)
  
def rand_by_correct(is_correct, preds, y_true):
    return rand_by_mask((preds == y_true) == is_correct)
```

```python
def plots(ims, figsize=(12,6), rows=1, titles=None):
    f = plt.figure(figsize=figsize)
    for i in range(len(ims)):
        sp = f.add_subplot(rows, len(ims)//rows, i+1)
        sp.axis('Off')
        if titles is not None: 
            sp.set_title(titles[i], fontsize=16)
        plt.imshow(ims[i])
```

```python
from imageio import imread


def load_img_id(idx):
    return imread(PATH/'subdataset/valid'/validation_generator.filenames[idx])

  
def plot_val_with_title(idxs, title):
    imgs=[]
    for i in idxs:
        imgs.append(load_img_id(i))
    
    title_probs = probs[idxs] 
    print(title)
    return plots(imgs, rows=1, titles=title_probs, figsize=(16,8))
```

```python
d = rand_by_correct(True, preds, validation_generator.classes)
```

```python
# 1. A few correct labels at random
tmp = rand_by_correct(True, preds, validation_generator.classes)
plot_val_with_title(tmp, "Correctly classified")
```

```python
# 2. A few incorrect labels at random
tmp = rand_by_correct(False, preds, validation_generator.classes)
plot_val_with_title(tmp, "Incorrectly classified")
```

```python
def most_by_mask(mask):
    idxs = np.where(mask)[0]
    return idxs[np.argsort(-probs[idxs])[:4]]

def most_by_correct(y, is_correct, preds, y_true): 
    return most_by_mask(((preds == y_true) == is_correct) & (y_true == y))
```

```python
plot_val_with_title(most_by_correct(0, True, preds, validation_generator.classes), "Most correct cats")
```

```python
plot_val_with_title(most_by_correct(1, True, preds, validation_generator.classes), "Most correct dogs")
```

```python
plot_val_with_title(most_by_correct(0, False, preds, validation_generator.classes), "Most incorrect cats")
```

```python
plot_val_with_title(most_by_correct(1, False, preds, validation_generator.classes), "Most incorrect dogs")
```

```python
most_uncertain = np.argsort(np.abs(probs -0.5))[:4]
plot_val_with_title(most_uncertain, "Most uncertain predictions")
```

## 5- Choosing a learning rate

The *learning rate* determines how quickly or how slowly you want to update the *weights* (or *parameters*). Learning rate is one of the most difficult parameters to set, because it significantly affects model performance.

![alt text](https://raw.githubusercontent.com/valeoai/tutorial-images/master/images/01_Classification_Introduction/learningrates.jpeg "CS231N - Babysitting the learning process")

There are some built-in functions implemented in keras to scheddule the learning rate (a good example [here](https://github.com/keras-team/keras/blob/master/examples/cifar10_resnet.py)). Indeed, as we have seen during the lesson, while we progress in the training process, it is good to reduce the learning rate. For that purpose, we can use a scheduler <kbd>LearningRateScheduler</kbd>. We can also adopt some different strategies in case the lost does not progress anymore. this is done using the function <kbd>ReduceLROnPlateau</kbd>

![alt text](https://raw.githubusercontent.com/valeoai/tutorial-images/master/images/01_Classification_Introduction/LR_finder.jpeg "Learning rate")

There is a good post [here](https://towardsdatascience.com/understanding-learning-rates-and-how-it-improves-performance-in-deep-learning-d0d4059c1c10) introducing some methods to deal with the learning rate .

The method `LR_Finder` helps you find an optimal learning rate. It uses the technique developed in the 2015 paper [Cyclical Learning Rates for Training Neural Networks](http://arxiv.org/abs/1506.01186), where we simply keep increasing the learning rate from a very small value, until the loss stops decreasing. We can plot the learning rate across batches to see what this looks like.

Few other readings [here](https://towardsdatascience.com/estimating-optimal-learning-rate-for-a-deep-neural-network-ce32f2556ce0) and [here](https://towardsdatascience.com/learning-rate-schedules-and-adaptive-learning-rate-methods-for-deep-learning-2c8f433990d1)

We first create a new model, since we want to know how to set the learning rate for a new (untrained) model.

```python
from keras.callbacks import Callback
from dlutils.lr_finder import LRFinder

model = Load_Model()
model.compile(optimizer='sgd',loss='binary_crossentropy', metrics=['accuracy'])
lr_finder = LRFinder(min_lr=1e-5, 
                     max_lr=1e-1, 
                     steps_per_epoch=np.ceil(train_generator.n/batch_size))

# Fit the model on the batches
model.fit_generator(train_generator,callbacks=[lr_finder])

lr_finder.plot_loss()
```

The loss is still clearly improving at lr=1e-2 (0.01), so that's what we use. Note that the optimal learning rate can change as we train the model, so you may want to re-run this function from time to time.

```python
from keras import optimizers

model = Load_Model()
sgd = optimizers.SGD(lr=0.01)
model.compile(optimizer=sgd,loss='binary_crossentropy', metrics=['accuracy'])

epochs = 4

history = model.fit_generator(train_generator, epochs=epochs,
                    validation_data=validation_generator,
                    verbose=1, use_multiprocessing=True, workers=8)

# list all data in history
print(history.history.keys())
# summarize history for accuracy
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

When we train such Deep Learning model, we have to be able to evaluate how good went the learning process or in some other, are we confident that we can deploy what we have trained at larger scale (like in a car now!)?

It appears that a model can look good looking to the accuracy printed during the learining process but actually it would be a pretty bad model when time becomes to deploy it on a broader test set. How could it be possible?

This comes from overfitting issue. But before explaining what is overfitting, we can have already a feeling of what happens when we consider how many parameters a model have. Indeed, models with a large number of free parameters can describe an amazingly wide range of phenomena and can fit any amount of data available. Even if such a model agrees well with the available data, that doesn’t make it a good model. It may be just that there’s enough freedom in the model due to these free parameters that it can describe almost any data set of the given size, without capturing any genuine insights into the underlying phenomenon. This generates a good fit to the current available data but the model fails to generalize the scenario and fails when we input a test data point, this is because our model has not been exposed to such situation.

And when we talk about Neural Networks we are playing with nearly lakhs! of parameters(can be more accordingly) and deep networks can have millions of parameters. So it is very likely that we will overfit our model at a point of time.

Overfitting is then defined by the fact that a model models the “training data” too well. Overfitting happens when a model learns the detail and noise in the training data to the extent that it negatively impacts the performance of the model on new data.

<img src="https://raw.githubusercontent.com/valeoai/tutorial-images/master/images/01_Classification_Introduction/Overfitting.png" width="60%">

As we can see on the curves above, we overfit when our model learn somehow the noise of our training set. Our model has enough degree of freedom in its parameters to do that but it is useless to learn noise and even worst as it will result in a bad generalization, that is to say our model will behave poorly on new data.

That being said, let s go back to our initial question which how good is our model, or in some other words, does our model overfit?

There are few numbers outputed during the training process, and looking to these numbers (accuray and loss) gives an indication of training. We have to compare Loss and accuray between training and validation set. Training Loss will be all the time better than validation Loss but it should remain in the same order of magnitude. <b>If validation Loss is 10 times worst than training Loss, it is then quite likely that we overfit.</b>

For the accuracy, we should rather consider how it progresses, so it is better to look to the history for both training and validation. Indeed, we will see later than accuracy can be better or worst on the validtaion set compare to the training set but there are some reasons behind. <b>However, we can be sure than is the training accuracy is decreasing while the validation accuracy is not progressing, then it is quite likely that our model starts to overfit.</b>

Then, what can we do to avoid/reduce overfitting? This is what we will see in the next lessons: 
- Add more data
- Use data augmentation
- Use architectures that generalize well
- Add regularization (dropout)
- Reduce architecture complexity
