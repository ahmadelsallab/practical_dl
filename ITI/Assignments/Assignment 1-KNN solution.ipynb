```
import cv2
import numpy as np

class NN:
  def __init__(self):
    pass

  def preprocess_img(self, image):
    image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    return cv2.resize(image_gray, (32, 32)).flatten()

  '''
  def get_features(self, image):
    image_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    hist = cv2.calcHist([image_hsv], [0, 1, 2], None, (8, 8, 8), [0, 180, 0, 256, 0, 256])
    return cv2.resize(hist, (32,32)).flatten()  
  '''
  def prepare_data(self, path):
    
    #listdir() returns a list containing the names of the entries in the directory given by the path.
    imagesList = os.listdir(path)
    
    X = []
    Y = []

    for file in os.listdir(path):

      image = cv2.imread(os.path.join(path, file))
      
      
      #X.append(self.get_features(self.preprocess_img(image)))
      X.append(self.preprocess_img(image))
      
      Y.append(file.split('.')[0])
    return np.array(X), np.array(Y)
  
  # Just memorize the data
  def train(self, X, Y):
    self.X_train = X
    self.Y_train = Y
    
  def predict(self, X):    
    
    Y_pred = []
    for x in list(X):
      distances = np.sum(np.abs(self.X_train - x), axis=1)
      y_pred_idx = np.argmin(distances)
      y_pred = self.Y_train[y_pred_idx]
      Y_pred.append(y_pred)
      
      # Y_pred.append(self.Y_train[np.argmin(np.sum(np.abs(self.X_train - x), axis=1))])
    return np.array(Y_pred)
```

```
train_path = 'train'

nn_cls = NN()
X, Y = nn_cls.prepare_data(train_path)
```

```


# Split into train and test
X_test = X[20000:,:]
Y_test = Y[20000:]
X = X[:20000,:]
Y = Y[:20000]




nn_cls.train(X, Y)

  
```

```
Y_pred = nn_cls.predict(X_test)
```
